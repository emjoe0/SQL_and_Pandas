```python
#Converting dataframe to an array of array
dfRows = df.values
dfRows
```

```python
#Converting dataframe from an array of array to a list of list
dfListRows = list(dfRows)
dfListRows
```

```python
#converting dataframe directly to a tuple of tuple. I noticed that converting from array of array (df.values) refused to convert
#It is very neccessary to convert the dataframe rows to a tuple to enable it inserted into the sql database
dfTupleRows = tuple(df.itertuples(index=False, name=None))
dfTupleRows
```

```python
#Inserting the tuple into the sql database using cur.executemany cursor to enable it accept both the query and the tuple
insertQuery = "insert into winequality_red values(?,?,?,?,?,?,?,?,?,?,?,?,?)"
cur.executemany(insertQuery, dfTupleRows)
```

```python
#Using pandas to read my new table in sql database
tableQuery = "select * from winequality_red"
cur.execute(tableQuery)
sqlDF = pd.read_sql_query(tableQuery, conn)
sqlDF
```

```python
subQuery = "select * from winequality_red where alcohol > 10 order by alcohol desc"
cur.execute(subQuery)
sqlDF2 = pd.read_sql_query(subQuery, conn)
sqlDF2
```
